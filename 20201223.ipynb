## 7.1.2 앤스콤 데이터 집합들로 그래프 만들기

```
# 앤스콤 데이터 집합은 4개의 데이터 그룹으로 구성되어 있으며, 각 데이터 그룹의 차이를 파악하려면 그래프로 시각화해야한다.
# 따라서 이번에는 모든 데이터 그룹을 이용하여 그래프를 그려볼 것이다.
```

```
# 1. 앤스콤 데이터프레임의 dataset 열의 값이 I,II,III,IV인 것을 불린 추출하여 dataset_1,2,3,4에 저장한다.
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

```
# 2. 그래프 격자가 위치할 기본 틀을 만든다.
fig = plt.figure()
```

## 7.2.1 기초 그래프 그리기

```
# seaborn 라이브러리에는 tips이라는 데이터 집합이 있다. 
# tips 데이터 집합은 어떤 식당에서 팁을 지불한 손님의 정보를 모아둔 것이다.
# 우리는 이 tips 데이터 집합을 이용해 다양한 그래프를 그려볼 것이다.
```

```
# tips 데이터 집합을 불러와 변수 tips에 저장한다.
# tips 데이터프레임은 지불금액, 팁, 성별, 흡연유무, 요일, 시간, 전체 인원 정보를 담고있다.
tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

```
# 데이터프레임의 열 데이터 분포와 빈도를 살펴보기 위해 '히스토그램'을 사용한다.
# 데이터프레임의 total_bill, tip등의 열을 변수라고 한다. 
# 변수를 하나만 사용한 그래프를 '일변량 그래프'라고 한다.
```

```
# 기본 틀(fig)을 마련하고 그래프 격자(axes1)를 구성한다.
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

```
# hist 메서드에 total_bill 열을 전달하여 히스토그램을 만든다.
# 이때 x축의 간격은 bins 인자값으로 조정할 수 있다.
axes1.hist(tips['total_bill'], bins = 10)
axes1.set_title('History of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
fig
```

```
# 이번에는 변수 2개를 사용하여 산점도 그래프, 즉 '이변량 그래프'를 그려볼 것이다.
```

```
# 산점도 그래프
## 기본 틀과 그래프 격자를 만든다.
## scatter 메서드 <- total_bill, tips 열을 전달
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```
# 박스 그래프(Box Plot) 그리기
```

```
# 박스 그래프는 이산형 변수와 연속형 변수를 함께 사용하는 그래프이다.
# 이산형 변수 = Female, Male과 같이 명확하게 구분되는 값
# 연속형 변수 = Tip과 같이 명확하게 셀 수 없는 범위의 값
```

```
# boxplot 메서드를 사용하여 박스 그래프를 그린다.
boxplot = plt.figure()
axes1  = boxplot.add_subplot(1,1,1)

axes1.boxplot(
    [tips[tips['sex'] == 'Female']['tip'],
     tips[tips['sex'] == 'Male']['tip']],
     labels = ['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

## 7.2.2 다변량 그래프 그리기

```
# 3개 이상의 변수를 사용한 그래프를 '다변량 그래프'라고 부른다.
# 앞에서는 지불금액(total_bill)과 팁(tip)만을 사용하여 산점도 그래프를 그렸다.
# 만약 여기에 성별을 추가하고, 식사 비용을 추가하고 싶다면 어떻게 해야할까?
# 점의 색상을 다르게 하고, 점의 크기를 다르게 하면 될 것이다.   -> 변수추가

# 즉, 이렇게 3개 이상의 변수를 사용하는 다변량 그래프는 적재적소에 맞는 그래프 요소를 추가하여 그래프를 표현한다.
```

```
# 앞서 실습한 산점도 그래프(scatter_plot)에 성별을 새 변수로 추가해보자. 이는 색상으로 구분하면 적당할 것이다.
# 하지만 Female, Male과 같은 문자열로는 색상을 지정할 수 없다. 따라서 0,1과 같은 정수를 사용해야한다.
# 따라서 각 문자열을 정수로 치환하는 함수를 만들어야한다.
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

```
# 이제 recode_sex 메서드가 반환한 값(0,1)을 데이터프레임에 sex_color 열을 새로 생성하여 추가해보자.
# 이때 sex열에 recode_sex 함수를 브로드캐스팅하기 위하여 apply 메서드를 사용했다.
tips['sex_color'] = tips['sex'].apply(recode_sex)
print(tips['sex_color'].head())
```

```
# 이제는 Female, Male 같은 문자열이 아닌 0,1같은 숫자로 표현되고 있음을 알 수 있다.
```

```
# 그래프를 그리는 김에 테이블당 인원 수(size)도 산점도 그래프에 추가해보자.이는 점의 크기로 표현하는 것이 적당하다.
# 다음은 scatter 메서드에 s,c 인자값으로 테이블당 인원 수와 성별의 치환값을 전달한 것이다.
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)    # 위치
axes1.scatter(
    x=tips['total_bill'],   # 지불금액
    y=tips['tip'],          # 팁 비용
    s=tips['size']*50,      # s는 점의 크기(size)
    c=tips['sex_color'],    # c는 점의 색상(color)
    alpha = 0.5)            # alpha는 투명도 / 투명도를 제어하여 겹치는 점을 구분할 수 있다.
axes1.set_title('Total Bill vs Tip colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
# plt.legend()
# axes1.set_legend()
```

# 7.3 seaborn 자유롭게 사용하기

```
# 이제는 seaborn 라이브러리 자체를 활용하여 그래프를 그려볼 것이다.
# seaborn을 활용하면 matplotlib보다 좀 더 화려한 그래프를 그릴 수 있다.
```

```
# seaborn 라이브러리는 sns라는 이름으로 줄여 사용한다.
import seaborn as sns
# 다시 tips 데이터 집합을 불러온다.
tips = sns.load_dataset('tips')
```

## 7.3.1 단변량 그래프 그리기

```
# 히스토그램 + 밀집도 그래프 <- subplots, distplot 메서드

## 먼저 subplots 메서드로 기본 틀을 만든다.
## 그 다음 distplot 메서드에 total_bill 열 데이터를 전달한다.
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```
# (밀집도 그래프 제외) 히스토그램

## distplot 메서드를 사용하면 히스토그램과 밀집도를 같이 그린다.
## 만약 밀집도 그래프를 제외하고 싶다면 kde 인자값을 False로 설정하면 된다.
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Title Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```
# 밀집도 그래프

# 밀집도 그래프만 나타내려면 hist 인자를 False로 지정하면 된다.
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```
# 히스토그램 + 밀집도 + rug 인자 = 양탄자(rug) 그래프

## 양탄자 그래프는 그래프 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한다.
hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```
# count 그래프 -> 빈도수

## count 그래프는 이산값을 나타낸 그래프이다.
## 다음은 countplot 메서드에 tips 데이터프레임의 day열 데이터를 넣어 count그래프를 그린 것이다.
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Cout of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

## 7.3.2 이변량 그래프 그리기

```
# 산점도 그래프 <- regplot 메서드

## seaborn 라이브러리로 산점도 그래프를 그리려면 regplot 메서드를 사용해야 한다.
## regplot 메서드를 사용하면 산점도 그래프와 회귀선을 함께 그릴 수 있다.
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and TIP')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```
## 회귀선을 제거하려면 fit_reg 인자를 False로 지정하면 된다.

ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```
# 산점도 그래프 + 히스토그램 <- joinplot 메서드

## 이번에는 산점도 그래프와 히스토그램을 한 번에 그려주는 joinplot 메서드를 사용해보자.
## x,y 인자에 원하는 열 이름을 지정하고 data 인자값으로 데이터프레임을 지정하면 그래프를 그릴 수 있다.

joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)   # 여기서 y는 제목의 위치값이다.
```

```
# 하지만 산점도 그래프는 점이 겹쳐 보일 경우 구분하기 어렵다는 단점이 있다.
# 이를 해결하기 위해 육각 그래프(hexbin)를 사용한다.
```

```
# 육각 그래프 + 히스토그램

## 육각 그래프는 특정 데이터의 개수가 많아지면 점점 진한 색으로 표현된다.
## kind 인자값을 "hex"으로 지정하여 표현한다.

hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize = 10, y=1.03)
```

```
# 이차원 밀집도 <- kdeplot 메서드
kde, ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],  # x축
                 data2=tips['tip'],   # y축
                 shade=True)  # 음영효과
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```
# 바 그래프
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```
# 박스 그래프 <- Boxplot 메서드 사용
## 최솟값, 1분위수, 중간값, 3분위수, 최댓값, 이상치 등 다양한 통계량 한 번에 표현

ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```
# PairGrid  <-  map_upper, map_lower 메서드 사용 
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)   # map_upper() = 대각선을 기준으로 위쪽에 그릴 그래프
pair_grid = pair_grid.map_lower(sns.kdeplot)   # map_lower() = 대각선을 기준으로 아래쪽에 그릴 그래프
pair_grid = pair_grid.map_diag(sns.kdeplot)    # map_diag() = 대각선을 중심으로 그래프를 그린다! 
plt.show()

## 요약된 지표를 보여주고자 할 때, 확실한 지표를 보여주고자 할 때 쓰인다.
```

```
## tips 데이터 집합을 사용하여 그래프를 그려보자.
## 먼저 FaceGrid에 데이터프레임과 그룹을 구분할 열 이름을 전달하여 객체를 생성한다.
## 그런 다음 객체(facet)의 map 메서드에 그래프의 종류(displot)와 x축으로 지정할 열 이름을 순서대로 전달한다.
## 참고로 distplot() = 히스토그램 + 밀집도 그래프 / rug() = 양탄자 그래프

facet = sns.FacetGrid(tips, col='time')
facet.map(sns.distplot, 'total_bill', rug=True)

## 이렇게 하면 히스토그램, 밀집도, 양탄자 그래프를 그릴 수 있다.
```

```
## 다음은 tips 데이터 집합의 day 열로 그룹을 구분하여 그래프를 그린 것이다.
facet = sns.FacetGrid(tips, col='day', hue='sex')   # hue 인자값으로 색상에 사용할 열 이름을 추가
facet = facet.map(plt.scatter, 'total_bill','tip')
facet = facet.add_legend()   # legend()는 범례표시
```

```
## 이번에는 더 많은 변수, 즉 time, smoke열을 추가하여 그래프를 그려볼 것이다.
## time, smoker열은 각각 2개의 값을 가지고 있다. (Dinner/Lunch, Yes/No) -> 따라서 2행 2열의 격자가 만들어 진다.

facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
facet.map(plt.scatter, 'total_bill', 'tip')

## col,row 인자값에 따라 smoker = Yes l time = Lunch와 같은 방법으로 그래프를 구분한다.
```
